### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.

# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
# Additionally, create a histogram for given data with a title, x-axis label, and a specific color.

# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
# Add a histogram as a separate plot.

import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

x = np.linspace(0, 10, 100)
y = x**2
axes[0, 0].plot(x, y)
axes[0, 0].set_title('Line Plot')
axes[0, 0].set_xlabel('X-axis')
axes[0, 0].set_ylabel('Y-axis')
axes[0, 0].grid(True)

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axes[0, 1].set_title('Bar Plot')

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
axes[1, 0].scatter(x, y, color='orange')
for i in range(len(x)):
    axes[1, 0].annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(5,-5))
axes[1, 0].set_title('Scatter Plot')

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['gold', 'blue', 'red', 'green'])
axes[1, 1].set_title('Pie Chart')

data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
axes[0, 2].hist(data, bins=5, color='cyan', edgecolor='black')
axes[0, 2].set_title('Histogram')
axes[0, 2].set_xlabel('Value')
axes[0, 2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
