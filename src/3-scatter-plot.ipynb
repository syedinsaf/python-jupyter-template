### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

assert isinstance(categories, list) and isinstance(values, list) and isinstance(colors, list)
assert len(categories) == len(values) == len(colors)

plt.figure(figsize=(6, 4))
plt.bar(categories, values, color=colors)

plt.title('Bar Plot of Categories and Values')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

assert isinstance(x, list) and isinstance(y, list)
assert len(x) == len(y)

plt.figure(figsize=(6, 4))
plt.scatter(x, y, color='blue')

for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(5, 5), ha='center')

plt.title('Scatter Plot with Annotations')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)

plt.show()
```
